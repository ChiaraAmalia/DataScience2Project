```python
# importare tutte le librerie necessarie
import pandas as pd
import numpy as np
import datetime as dt
from tabulate import tabulate
import matplotlib.pyplot as plt
path = r".\\DataSet\\"
import seaborn as sns
```

```python
allMadrid = pd.read_csv(path+"out.csv")
allMadrid.head()
```

```python
# Faccio un cambio di formato sul campo data passando da stringa a data
allMadrid['date'] = pd.to_datetime(allMadrid['date'], format='%Y/%m/%d %H:%M:%S')
```

```python
#prendo tutti i valori del NO2 del 2018
NO2_in = allMadrid['NO_2']
BEN_in = allMadrid['BEN']
CO_in = allMadrid['CO']
EBE_in = allMadrid['EBE']
NMHC_in = allMadrid['NMHC']
NO_in = allMadrid['NO']
O3_in = allMadrid['O_3']
PM10_in = allMadrid['PM10']
PM25_in = allMadrid['PM25']
SO2_in = allMadrid['SO_2']
TCH_in = allMadrid['TCH']
TOL_in = allMadrid['TOL']
Date_in = allMadrid['date']
year_in = pd.DatetimeIndex(Date_in).year
month_in = pd.DatetimeIndex(Date_in).month
TAB = pd.DataFrame({'Date':Date_in, 'NO2':NO2_in,
                    'BEN':BEN_in,
                    'CO':CO_in,
                    'EBE':EBE_in,
                    'NMHC':NMHC_in,
                    'NO':NO_in,
                    'O3':O3_in,
                    'PM10':PM10_in,
                    'PM25':PM25_in,
                    'SO2':SO2_in,
                    'TCH':TCH_in,
                    'TOL':TOL_in,
                    'Month':month_in, 'Year':year_in})

AVG_YEAR=TAB.groupby(['Year']).mean()
TAB_avg = TAB.groupby(['Year','Month']).mean()
AVG_2018 = TAB_avg.filter(like='2018', axis=0)
AVG_2017 = TAB_avg.filter(like='2017', axis=0)
AVG_2016 = TAB_avg.filter(like='2016', axis=0)
AVG_2015 = TAB_avg.filter(like='2015', axis=0)
AVG_2014 = TAB_avg.filter(like='2014', axis=0)
AVG_2013 = TAB_avg.filter(like='2013', axis=0)
AVG_2012 = TAB_avg.filter(like='2012', axis=0)
AVG_2011 = TAB_avg.filter(like='2011', axis=0)
AVG_2010 = TAB_avg.filter(like='2010', axis=0)
AVG_2009 = TAB_avg.filter(like='2009', axis=0)
AVG_2008 = TAB_avg.filter(like='2008', axis=0)

display(TAB)
display(TAB_avg)
display(AVG_YEAR)
```
